```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np
%matplotlib inline
```

```python
#uploading the csv and deliminating data by commas since the data set is deliminated by TAB
#getting information on the dataset
marketing = pd.read_csv('marketing_campaign.csv', sep='\t', lineterminator='\r')
marketing.describe()
```

```python
marketing.info()
marketing.dropna(inplace=True)
marketing['Year_Birth'] = marketing['Year_Birth'].astype(int)
```

```python
marketing.info()
```

```python
marketing['Year_Birth'].unique()
```

Need to remove birth years 1893 and 1899, those are not logical based on the time period the customer would be over 100 years old.

```python
marketing.drop(marketing[marketing['Year_Birth'] < 1940].index, inplace = True)
```

```python
marketing['Year_Birth'].unique()
```

```python
marketing[marketing['Year_Birth'].isna()]
```

```python
#looking at the column headers to make sure the data in the csv is deliminated by commas
marketing.head()
```

```python
#creating the total sales column in the dataset for each customer ID for customer segmentation using IQR method
marketing['TotalDollarSales'] = marketing['MntWines'] + marketing['MntFruits'] + marketing['MntMeatProducts'] + marketing['MntFishProducts'] + marketing['MntSweetProducts'] + marketing['MntGoldProds']
print(marketing['TotalDollarSales'])
```

```python
#making sure the TotalDollarSales column is present
print(marketing.columns)
```

```python
#calculating the IQR of sales to begin segmentation
percentiles_sales=np.array([25,50,75])
ptiles_sales=np.percentile(marketing['TotalDollarSales'],[25,50,75])
print(ptiles_sales)
```

```python
#Generation Segmentation
marketing.loc[marketing['Year_Birth']<1965,'Generations']='Boomer'
marketing.loc[(marketing['Year_Birth']>=1965)&(marketing['Year_Birth']<1981),'Generations']='Gen_X'
marketing.loc[(marketing['Year_Birth']>=1981)&(marketing['Year_Birth']<1997),'Generations']='Millennials'
marketing.loc[marketing['Year_Birth']>=1997,'Generations']='Gen_Z'
```
